```python
# dataset의 자치구코드와 자치구명만 추출하여 중복 제거
dataset_unique_df = df_dataset[['자치구코드', '자치구명']].drop_duplicates().astype(str)

# 법정동 코드 목록에서 자치구코드와 자치구명 추출하여 중복 제거
answer_unique_df = df_answer[['자치구코드', '자치구명']].drop_duplicates().astype(str)

# 두 데이터셋을 자치구코드를 기준으로 병합
merged_df = dataset_unique_df.merge(answer_unique_df, on='자치구코드', how='left', suffixes=('', '_law'))

# 자치구명이 다른 경우 필터링
mismatched_df = merged_df[merged_df['자치구명'] != merged_df['자치구명_law']]

# 자치구명이 잘못 연결된 행의 자치구코드 목록 추출
mismatched_codes = mismatched_df['자치구코드'].unique()

# df_dataset에서 해당 자치구코드에 해당하는 모든 행 필터링
mismatched_rows_df = df_dataset[df_dataset['자치구코드'].astype(str).isin(mismatched_codes)]

#정답은?
correct_answer_df = df_answer[df_answer["자치구코드"].astype(str) == "26230"]
print( correct_answer_df) # 없음 ==> 부산이라서 그럼.
mismatched_rows_df
```

```python
# dataset_df에서 해당 자치구코드에 해당하는 모든 행 제거
cleaned_dataset_df = df_dataset.drop(101250)

print(cleaned_dataset_df.shape[0]) 
cleaned_dataset_df.head()
```

```python
# 자치구코드와 법정동코드를 합친 새로운 컬럼 생성
cleaned_dataset_df['자치구법정동코드'] = cleaned_dataset_df['자치구코드'].astype(str) + cleaned_dataset_df['법정동코드'].astype(str)
df_answer['자치구법정동코드'] = df_answer['자치구코드'].astype(str) + df_answer['법정동코드'].astype(str)

# 새로운 코드 기준으로 병합
merged_df = cleaned_dataset_df.merge(df_answer, on='자치구법정동코드', how='left', suffixes=('', '_law'))
merged_df
```

```python
# 자치구법정동코드 기준으로 법정동명이 다른 경우 필터링
mismatched_df = merged_df[merged_df['법정동명'] != merged_df['동명']]
print(mismatched_df.shape[0])
mismatched_df.to_excel("자치구법정동코드!=법정동명.xlsx")
mismatched_df.head()
```

```python
# 삭제하기
# 자치구법정동코드 기준으로 법정동명이 다른 경우 삭제
merged_df = merged_df.drop(mismatched_df.index)
```
