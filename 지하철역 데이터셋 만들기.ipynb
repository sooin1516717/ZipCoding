```python
# 1️중복된 역사명(환승역) 찾기
duplicate_stations = subway['역사명'].value_counts()
duplicate_stations = duplicate_stations[duplicate_stations >= 2].index  # 2회 이상 등장하는 역사명만 선택

# 2️환승역별 최적의 행 선택
selected_rows = []

for station in duplicate_stations:
    # 해당 역사명의 모든 데이터 가져오기
    station_data = subway[subway['역사명'] == station].copy()
    
    # 환승노선번호를 문자열(str)로 변환
    station_data['환승노선번호'] = station_data['환승노선번호'].astype(str)

    # 데이터기준일자(`datetime`) 변환
    station_data['데이터기준일자'] = pd.to_datetime(station_data['데이터기준일자'])

    # 환승노선번호가 가장 긴 (환승역수가 가장 긴) 행 선택
    station_data = station_data.sort_values(by='환승노선번호', key=lambda x: x.str.len(), ascending=False)

    # 데이터 기준일이 최신인 행 선택 (환승노선 개수가 같을 경우)
    best_station = station_data.sort_values(by='데이터기준일자', ascending=False).iloc[0]
    
    # 최적의 행 저장
    selected_rows.append(best_station)

# 최적의 환승역 데이터프레임 생성
optimized_subway = pd.DataFrame(selected_rows)

# value_counts() 값이 1인 역사들만 필터링
single_station_subway = subway[~subway['역사명'].isin(duplicate_stations)]

# 최종 데이터프레임 생성 (환승역 최적화 + 단일 역 유지)
subway = pd.concat([optimized_subway, single_station_subway], ignore_index=True)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderUnavailable
import time

# Geopy Geocoder 설정 (Nominatim: OpenStreetMap 기반)
geolocator = Nominatim(user_agent="subway_locator", timeout=10)

# 주소 → 위도/경도 변환 함수 (재시도 기능 추가)
def get_lat_lon_from_address(address, retries=3):
    """
    Geopy를 사용하여 주소를 위도/경도로 변환하는 함수 (재시도 포함)
    """
    for attempt in range(retries):
        try:
            location = geolocator.geocode(address)
            if location:
                return location.latitude, location.longitude
        except (GeocoderTimedOut, GeocoderUnavailable) as e:
            print(f"⏳ 재시도 {attempt + 1}/{retries} - {address} (에러: {e})")
            time.sleep(2)  # 재시도 전 대기

    return None, None  # 좌표를 찾을 수 없는 경우 None 반환

# NaN 값을 채울 딕셔너리 (API 중복 요청 방지)
address_cache = {}

# 역위도 / 역경도가 NaN인 경우 주소 기반으로 채우기
for index, row in subway.iterrows():
    if pd.isna(row["역위도"]) or pd.isna(row["역경도"]):  # NaN 값이 있는 경우
        address = row["역사도로명주소"]

        # 캐시에 주소가 없으면 API 요청
        if address not in address_cache:
            lat, lon = get_lat_lon_from_address(address)
            address_cache[address] = (lat, lon)  # 캐싱
            time.sleep(1)  # 요청 간격 조절 (초당 1회 제한)
        else:
            lat, lon = address_cache[address]  # 캐시된 값 사용

        # DataFrame 업데이트
        subway.at[index, "역위도"] = lat
        subway.at[index, "역경도"] = lon
```

```python
import requests
import pandas as pd
import time
import re

# ✅ Kakao API Key 설정 (사용자 API 키 입력)
KAKAO_API_KEY = "929ad4c932b9aedfb9592f91777ffa6a"

# ✅ Kakao 주소 검색 함수
def get_lat_lon_kakao(address):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {"query": address}

    response = requests.get(url, headers=headers, params=params)
    result = response.json()

    if "documents" in result and len(result["documents"]) > 0:
        return float(result["documents"][0]["y"]), float(result["documents"][0]["x"])
    
    return None, None  # 🔹 응답이 없을 경우 None 반환

# ✅ 도로명 주소 전처리 함수
def preprocess_address(address):
    # 🔹 괄호 안의 정보 제거 (ex. "서울특별시 마포구 마포대로 지하100(공덕동)" → "서울특별시 마포구 마포대로 지하100")
    address = re.sub(r"\(.*?\)", "", address)
    
    # 🔹 '지하' 제거 후 공백 정리 (ex. "서울특별시 강동구 명일로 지하 117" → "서울특별시 강동구 명일로 117")
    address = address.replace("지하", "").strip()
    
    return address

# ✅ NaN 데이터를 채울 딕셔너리 (API 중복 요청 방지)
address_cache = {}

# ✅ 역위도 / 역경도가 NaN인 경우 주소 기반으로 채우기
for index, row in subway.iterrows():
    if pd.isna(row["역위도"]) or pd.isna(row["역경도"]):  # NaN 값이 있는 경우
        original_address = row["역사도로명주소"]
        cleaned_address = preprocess_address(original_address)  # 주소 전처리

        # 🔹 캐시에 주소가 없으면 API 요청
        if cleaned_address not in address_cache:
            lat, lon = get_lat_lon_kakao(cleaned_address)
            address_cache[cleaned_address] = (lat, lon)  # 캐싱
            time.sleep(0.1)  # Kakao API 속도 조절
        else:
            lat, lon = address_cache[cleaned_address]  # 캐시된 값 사용

        # 🔹 DataFrame 업데이트
        subway.at[index, "역위도"] = lat
        subway.at[index, "역경도"] = lon

# ✅ 남은 NaN 개수 확인
missing_values = subway[subway['역위도'].isnull() | subway['역경도'].isnull()]
print(f"📌 아직 남은 NaN 개수: {missing_values.shape[0]}")
```

```python
import pandas as pd

# `seoul_subway`에서 필요한 컬럼만 추출
seoul_subway = seoul_subway[['역사명', '호선', '위도', '경도']].copy()

# `subway`에서 필요한 컬럼만 추출 (노선명 컬럼 존재 확인)
if '노선명' not in subway.columns:
    raise KeyError("🚨 subway 데이터프레임에 '노선명' 컬럼이 없습니다. 컬럼명을 확인해주세요.")

# `seoul_subway`에서 `역사명`과 `subway['노선명']` 기준으로 중복 제거
def filter_seoul_subway(subway_row, seoul_subway_df):
    """ 
    subway의 `노선명`과 동일한 `호선`이 존재하는 경우 해당 행 유지
    없을 경우 해당 역사명에서 첫 번째 값 선택
    """
    matched_rows = seoul_subway_df[
        (seoul_subway_df['역사명'] == subway_row['역사명']) &
        (seoul_subway_df['호선'] == subway_row['노선명'])
    ]

    # 노선명이 일치하는 행이 있으면 첫 번째 값 반환
    if not matched_rows.empty:
        return matched_rows.iloc[0]

    # 동일한 역사명이 `seoul_subway`에 존재하는지 체크 후 반환
    all_matched_rows = seoul_subway_df[seoul_subway_df['역사명'] == subway_row['역사명']]
    
    if not all_matched_rows.empty:
        return all_matched_rows.iloc[0]
    
    # 역사명이 `seoul_subway`에도 없으면 None 반환 (이후 fillna 처리)
    return pd.Series({'역사명': subway_row['역사명'], '위도': None, '경도': None})

# `subway` 데이터프레임의 `역사명`을 기준으로 `seoul_subway` 필터링 적용
filtered_seoul_subway = subway.apply(lambda row: filter_seoul_subway(row, seoul_subway), axis=1)

# 필요한 컬럼만 유지
filtered_seoul_subway = filtered_seoul_subway[['역사명', '위도', '경도']]

# `subway`와 `filtered_seoul_subway`를 `역사명` 기준으로 병합
merged_df = subway.merge(filtered_seoul_subway, on='역사명', how='left', suffixes=('', '_seoul'))

# 병합 후 `역위도`, `역경도`가 NaN인 경우 `seoul_subway`의 `위도`, `경도` 값으로 채우기
merged_df['역위도'] = merged_df['역위도'].fillna(merged_df['위도'])
merged_df['역경도'] = merged_df['역경도'].fillna(merged_df['경도'])

# 불필요한 보조 컬럼 삭제
merged_df.drop(columns=['위도', '경도'], inplace=True)

# 중복 확인 (혹시 남아있는지 확인)
print(f"📌 중복된 역사명 개수: {merged_df['역사명'].duplicated().sum()}")

# 남은 NaN 개수 확인
missing_values = merged_df[merged_df['역위도'].isnull() | merged_df['역경도'].isnull()]
print(f"📌 아직 남은 NaN 개수: {missing_values.shape[0]}")

```
