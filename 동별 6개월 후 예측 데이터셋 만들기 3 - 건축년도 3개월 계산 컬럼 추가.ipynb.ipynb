```python
df = pd.read_pickle("C:/ZIPcoding/df_after_fill_NaN.pkl")
```

```python
df.info()
```

```python
df2019 = pd.read_csv("C:/ZIPcoding/original_data/서울시 부동산 실거래가 정보 2019.csv", encoding = 'euc=kr')
```

```python
df = df[['자치구코드', '법정동코드', '계약일', '건축년도']]
```

```python
df.info()
```

```python
df2019 = df2019[['자치구코드', '법정동코드', '계약일', '건축년도','건물용도']]
```

```python
df2019['계약일'] = pd.to_datetime(df2019['계약일'].astype(str), format='%Y%m%d')
```

```python
df2019["건축년도"] = df2019.groupby(["자치구코드", "법정동코드", "건물용도"])["건축년도"].transform(lambda x: x.fillna(x.median()))
```

```python
df2019["건축년도"] = df2019.groupby(["자치구코드", "법정동코드"])["건축년도"].transform(lambda x: x.fillna(x.median()))
```

```python
df2019['건축년도'] = df2019['건축년도'].astype(int)
```

```python
df2019 = df2019.iloc[:,:4]
```

```python
df2019.shape
```

```python
for_year = pd.concat([df, df2019])
```

```python
for_year
```

```python
df_grouped = df.groupby(['자치구코드', '법정동코드', '계약일', '건물용도']).size().unstack(fill_value=0).reset_index()
```

```python
dong_df = pd.merge(dong_df, df_grouped, on = ['자치구코드', '법정동코드', '계약일'], how='left')
```

```python
for_year = for_year[for_year['계약일'] >= '2020-01-01']
```

```python
for_year = for_year.groupby(['자치구코드', '법정동코드'])['건축년도'].agg(['mean', 'median', 'std']).reset_index()
```

```python
for_year.columns = ['자치구코드', '법정동코드', '건축년도평균', '건축년도중앙값', '건축년도표준편차']
```

```python
dong_df = pd.merge(dong_df, for_year, on = ['자치구코드', '법정동코드'], how='left')
```
