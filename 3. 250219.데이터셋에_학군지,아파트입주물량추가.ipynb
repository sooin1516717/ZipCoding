```python
import pandas as pd

csv_file_path = "기준금리_대출금리_추가_데이터(2020-2025).csv"
csv_file_school = "서울5대학군지.csv"
csv_file_apart = "수도권_전국아파트_입주물량(2020-2025).csv"

df_dataset = pd.read_csv(csv_file_path, low_memory=False)
df_school = pd.read_csv(csv_file_school, low_memory=False)
df_apart = pd.read_csv(csv_file_apart, low_memory=False)


print(f"총 행 수: {df_dataset.shape[0]}")

# '법정동명' 컬럼에서 결측값 확인
missing_values1 = df_dataset["법정동명"].isnull().sum()

# '계약연월' 컬럼에서 결측값 확인
missing_values2 = df_dataset["계약연월"].isnull().sum()

# 결과 출력
print(f" 컬럼의 결측값 개수: {missing_values1, missing_values2}\n")

print(df_dataset.head())
df_school.head()
```

```python
# "자치구명, 법정동명 "을 기준으로 맵핑. 서울5대학군지여부 컬럼 추가
merged_df = pd.merge(df_dataset, df_school, on =['자치구명', '법정동명'], how='left')

# nan 값은 0으로 체체
merged_df['서울5대학군지여부'] = merged_df['서울5대학군지여부'].fillna(0)
merged_df.head()
```

```python
# "계약연월 "을 기준으로 맵핑. 아파트입주물량 건수 컬럼 추가됨
merged_df = pd.merge(merged_df, df_apart, left_on=['계약연월'], right_on=['연월'], how='left')

merged_df.head()
```

```python
# 불필요한 컬럼들 제거
merged_df = merged_df.drop(columns=["연월","연도","월"])
merged_df.head()
```

```python
missing_values = merged_df["전국아파트입주물량(수도권)"].isnull().sum()
print(f"'전국아파트입주물량' 컬럼의 결측값 개수: {missing_values}")
print(f"총 행 수: {merged_df.shape[0]}")
```

```python
merged_df.to_csv("금리,학군지,아파트물량_추가_데이터(2020-2025).csv", index=False)
```
