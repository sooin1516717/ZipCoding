- 날씬이 지하철 정보를 뚱땡이 본번 부번에 있는 거와 맵핑하기

```python
import pandas as pd

csv_data = "dataset0220_suin_2.csv"
csv_subway = "address_with_hangang_subway_jjin.csv"

df_data = pd.read_csv(csv_data, low_memory=False)
df_subway = pd.read_csv(csv_subway, low_memory=False)

print(f"총 행 수: {df_subway.shape[0]}")
print( df_subway.isna().sum().sort_values(ascending=True)) 
df_subway.head()
```

```python
df_data.head()
```

```python
# 매핑을 위한 키 생성 (자치구명, 법정동명, 본번, 부번을 하나의 문자열로 결합)
df_data['매핑키'] = (
    df_data['자치구코드'].astype(str) + "_" +
    df_data['법정동코드'].astype(str) + "_" +
    df_data['본번'].astype(str) + "_" +
    df_data['부번'].astype(str)
)

df_subway['매핑키'] = (
    df_subway['자치구코드'].astype(str) + "_" +
    df_subway['법정동코드'].astype(str) + "_" +
    df_subway['본번'].astype(str) + "_" +
    df_subway['부번'].astype(str)
)

# 두 데이터프레임에 동일한 '매핑키' 컬럼이 생성된 상태에서 조인 수행
df_merged = df_data.merge(df_subway, on='매핑키', how='left', suffixes=('', '_subway'))
df_merged.head()
```

```python
# 불필요한 컬럼들 제거
df_merged = df_merged.drop(columns=["매핑키","자치구코드_subway","법정동코드_subway", "본번_subway", "부번_subway"])
df_merged.head()
```

```python
print(f"총 행 수: {df_merged.shape[0]}")
```
