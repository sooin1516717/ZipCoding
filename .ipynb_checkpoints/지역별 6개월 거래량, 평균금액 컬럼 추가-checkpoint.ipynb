### ìì¹˜êµ¬, ë²•ì •ë™ë³„ 6ê°œì›” ê±°ë˜ëŸ‰, í‰ê·  ê±°ë˜ê¸ˆì•¡ ê³„ì‚°í•´ì„œ íŒŒì¼ ì €ì¥

```python
import pandas as pd
import mysql.connector

for i in range(2020, 2026):
    # âœ… MySQL ì—°ê²° ì„¤ì •
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="toor",
        database="zipcoding"
    )
    
    query = f"""
       WITH ê±°ë˜ëŸ‰_í‰ê·  AS (
        SELECT
            ì¸ë±ìŠ¤,
            ê³„ì•½ì¼,
            ë¬¼ê±´ê¸ˆì•¡,
            ìì¹˜êµ¬ì½”ë“œ,
            ë²•ì •ë™ì½”ë“œ,
    
            -- âœ… ìì¹˜êµ¬ ê¸°ì¤€ ìµœê·¼ 6ê°œì›” ê±°ë˜ëŸ‰
            COUNT(ë¬¼ê±´ê¸ˆì•¡) OVER (
                PARTITION BY ìì¹˜êµ¬ì½”ë“œ
                ORDER BY ê³„ì•½ì¼ 
                RANGE BETWEEN INTERVAL 6 MONTH PRECEDING AND CURRENT ROW
            ) AS ìì¹˜êµ¬ìµœê·¼6ê°œì›”ê±°ë˜ëŸ‰,
    
            -- âœ… ìì¹˜êµ¬ ê¸°ì¤€ ìµœê·¼ 6ê°œì›” í‰ê·  ë¬¼ê±´ê¸ˆì•¡
            AVG(ë¬¼ê±´ê¸ˆì•¡) OVER (
                PARTITION BY ìì¹˜êµ¬ì½”ë“œ
                ORDER BY ê³„ì•½ì¼ 
                RANGE BETWEEN INTERVAL 6 MONTH PRECEDING AND CURRENT ROW
            ) AS ìì¹˜êµ¬ìµœê·¼6ê°œì›”í‰ê· ê¸ˆì•¡,
    
            -- âœ… ìì¹˜êµ¬ + ë²•ì •ë™ ê¸°ì¤€ ìµœê·¼ 6ê°œì›” ê±°ë˜ëŸ‰
            COUNT(ë¬¼ê±´ê¸ˆì•¡) OVER (
                PARTITION BY ìì¹˜êµ¬ì½”ë“œ, ë²•ì •ë™ì½”ë“œ
                ORDER BY ê³„ì•½ì¼ 
                RANGE BETWEEN INTERVAL 6 MONTH PRECEDING AND CURRENT ROW
            ) AS ë™ìµœê·¼6ê°œì›”ê±°ë˜ëŸ‰,
    
            -- âœ… ìì¹˜êµ¬ + ë²•ì •ë™ ê¸°ì¤€ ìµœê·¼ 6ê°œì›” í‰ê·  ë¬¼ê±´ê¸ˆì•¡
            AVG(ë¬¼ê±´ê¸ˆì•¡) OVER (
                PARTITION BY ìì¹˜êµ¬ì½”ë“œ, ë²•ì •ë™ì½”ë“œ
                ORDER BY ê³„ì•½ì¼ 
                RANGE BETWEEN INTERVAL 6 MONTH PRECEDING AND CURRENT ROW
            ) AS ë™ìµœê·¼6ê°œì›”í‰ê· ê¸ˆì•¡
    
        FROM original_data
    )
    SELECT * FROM ê±°ë˜ëŸ‰_í‰ê· 
    WHERE YEAR(ê³„ì•½ì¼) = {i}  
    AND ë™ìµœê·¼6ê°œì›”ê±°ë˜ëŸ‰ IS NOT NULL  -- âœ… ë²•ì •ë™ ë°ì´í„°ì—ì„œ ê±°ë˜ê°€ ìˆëŠ” í–‰ë§Œ ìœ ì§€
    ORDER BY ê³„ì•½ì¼ DESC;
    
        
    """
    
    try:
        # âœ… MySQL ë°ì´í„° ì¡°íšŒ í›„ DataFrameìœ¼ë¡œ ë³€í™˜
        df = pd.read_sql(query, conn)
    
        # âœ… CSVë¡œ ì €ì¥
        df.to_csv(f"C:/ZIPcoding/region_output_{i}.csv", index=False, encoding="utf-8-sig")
    
        # âœ… Excel íŒŒì¼ë¡œ ì €ì¥ (ì„ íƒ)
        # df.to_excel("C:/ZIPcoding/output_2020.xlsx", index=False, encoding="utf-8-sig")
    
        print("âœ… ë°ì´í„° ì €ì¥ ì™„ë£Œ!")
    
    except Exception as e:
        print(f"ğŸš¨ ì˜¤ë¥˜ ë°œìƒ: {e}")
    
    finally:
        # âœ… MySQL ì—°ê²° ì¢…ë£Œ
        conn.close()
        print("âœ… MySQL ì—°ê²° ì¢…ë£Œë¨")
```
