```python
import pandas as pd

interest_rate_path = "기준금리_데이터.csv"
df_interest_rate = pd.read_csv(interest_rate_path, encoding="utf-8", low_memory=False)

print(df_dataset['계약일'].dtype)
```

```python
# YYYYMMDD 형식으로 되어 있다면, 이를 문자열로 변환한 후 to_datetime()을 적용합니다.
# 날짜 변환 (df_dataset)
df_dataset['계약일(날짜)'] = df_dataset['계약일'].astype(str).str.replace(r'[^\d]', '', regex=True)
df_dataset['계약일(날짜)'] = pd.to_datetime(df_dataset['계약일(날짜)'], format='%Y%m%d', errors='coerce')
print(df_dataset[['계약일(날짜)']])

# 날짜 변환 (df_interest_rate)
df_interest_rate['계정항목(날짜)'] = pd.to_datetime(df_interest_rate['계정항목'], errors='coerce')
df_interest_rate
```

```python
# 데이터 정렬 (merge_asof 사용을 위해 필수)
df_dataset = df_dataset.sort_values(by="계약일(날짜)")
df_interest_rate = df_interest_rate.sort_values(by="계정항목(날짜)")

print(df_dataset.head())
df_interest_rate.head()
```

```python
# 가장 가까운 과거 날짜 기준으로 병합
df_merged = pd.merge_asof(df_dataset, df_interest_rate, left_on="계약일(날짜)", right_on="계정항목(날짜)", direction="backward")

# 불필요한 "계정항목" 컬럼 제거
df_merged = df_merged.drop(columns=["계정항목"])

df_merged.to_csv("금리추가_데이터(2020-2025).csv", index=False)
df_merged.head()
```

```python
# 계약일을 날짜 형식으로 변환 (숫자로 저장된 경우 대비)
df_dataset["계약연월"] = pd.to_datetime(df_dataset["계약일"].astype(str), format="%Y%m%d", errors="coerce")

# 계약연월 컬럼 생성 (YYYYMM 형식)
df_dataset["계약연월"] = df_dataset["계약연월"].dt.strftime("%Y%m")
```

```python
# 금리 데이터 가공 (전치 및 컬럼 정리)
import pandas as pd

merged_file_path = "금리추가_데이터(2020-2025).csv"
csv_file_path = "예금은행 대출금리(신규취급액 기준).csv" 

df_csv = pd.read_csv(csv_file_path, encoding="utf-8", low_memory=False)
df_merged = pd.read_csv(merged_file_path, encoding="utf-8", low_memory=False)

df_trans = df_csv.T  # 행과 열을 바꾸기
df_trans = df_trans.reset_index()  # 인덱스를 컬럼으로 변환
df_trans

df_trans.columns = df_trans.iloc[1]  # 첫 번째 행을 컬럼명으로 설정
df_trans = df_trans[4:].reset_index(drop=True)  # 첫 번째 행 제거 후 인덱스 초기화
df_trans.columns = df_trans.columns.str.replace("3)", "", regex=False)
df_trans.head()
```

```python
# 📌 "계정항목"을 문자열로 변환 후, 날짜 변환하여 "계정연월" 생성
df_trans["계정연월"] = pd.to_datetime(df_trans["계정항목"].astype(str), errors="coerce").dt.strftime("%Y%m")

# 📌 "계약일"을 변환하여 "계약연월" 생성 (YYYYMM 형식)
df_merged["계약연월"] = pd.to_datetime(df_merged["계약일"].astype(str), format="%Y%m%d", errors="coerce")
df_merged["계약연월"] = df_merged["계약연월"].dt.strftime("%Y%m")

# 📌 astype(int)를 사용해 문자열을 숫자로 변환하여 merge_asof가 동작하도록 함.
df_merged["계약연월"] = df_merged["계약연월"].astype(int)
df_trans["계정연월"] = df_trans["계정연월"].astype(int)

print(df_trans.head())
df_merged.head()
```

```python
# 📌 "계약연월"을 기준으로 "계정연월"과 매칭하여 "고정형 주택담보대출", "변동형 주택담보대출" 컬럼 추가
# df_merged = df_merged.merge(df_trans[["계정연월", "고정형 주택담보대출", "변동형 주택담보대출"]], 
#                             left_on="계약연월", 
#                             right_on="계정연월", 
#                             how="left")

df_merged = df_merged.sort_values(by="계약일(날짜)")
df_trans = df_trans.sort_values(by="계정연월")

df_merged = pd.merge_asof(df_merged, df_trans, left_on="계약연월", right_on="계정연월", direction="backward")
df_merged
```

```python
# 행 수 출력
print(f"총 행 수: {df_merged.shape[0]}")
```

```python
# 불필요한 "계정항목" , 계정연월 컬럼 제거
df_merged = df_merged.drop(columns=["계정항목"])
df_merged = df_merged.drop(columns=["계정연월"])
```
