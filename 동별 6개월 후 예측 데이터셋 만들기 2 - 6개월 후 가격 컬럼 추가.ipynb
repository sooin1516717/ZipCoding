```python
import pandas as pd

day_offset = 1
while merged_df[['평당가격_6개월후', '실거래가_6개월후']].isna().any().any():
    # ✅ NaN 개수 확인 (반복문 시작 전)
    before_nan_price = merged_df['평당가격_6개월후'].isna().sum()
    before_nan_deal = merged_df['실거래가_6개월후'].isna().sum()
    
    if day_offset % 5 == 0 or before_nan_price > 0 or before_nan_deal > 0:
        print(f"🚀 NaN 채우기 진행 중... (-{day_offset}일 데이터 사용)")
        print(f"   ▶ 채우기 전: 평당가격 NaN = {before_nan_price}, 실거래가 NaN = {before_nan_deal}")

    # ✅ 하루 전 날짜 계산
    merged_df[f'계약일_6개월후_minus_{day_offset}'] = merged_df['계약일_6개월후'] - pd.DateOffset(days=day_offset)

    # ✅ 하루 전 날짜를 기준으로 미래 가격 가져오기 (법정동 기준)
    merged_df = pd.merge(merged_df, future_prices, 
                         left_on=['자치구코드', '법정동코드', f'계약일_6개월후_minus_{day_offset}'],
                         right_on=['자치구코드', '법정동코드', '계약일'], how='left', suffixes=('', f'_minus_{day_offset}'))

    # ✅ NaN 값 채우기 (법정동 기준)
    merged_df['평당가격_6개월후'].fillna(merged_df.get(f'평당가격_6개월후_minus_{day_offset}', pd.NA), inplace=True)
    merged_df['실거래가_6개월후'].fillna(merged_df.get(f'실거래가_6개월후_minus_{day_offset}', pd.NA), inplace=True)

    # ✅ 같은 자치구의 하루 전 가격도 가져오기
    merged_df = pd.merge(merged_df, future_prices_gu, 
                         left_on=['자치구코드', f'계약일_6개월후_minus_{day_offset}'],
                         right_on=['자치구코드', '계약일'], how='left', suffixes=('', f'_gu_minus_{day_offset}'))

    # ✅ NaN 값 채우기 (자치구 기준)
    merged_df['평당가격_6개월후'].fillna(merged_df.get(f'평당가격_6개월후_gu_minus_{day_offset}', pd.NA), inplace=True)
    merged_df['실거래가_6개월후'].fillna(merged_df.get(f'실거래가_6개월후_gu_minus_{day_offset}', pd.NA), inplace=True)

    # ✅ 불필요한 컬럼 삭제
    merged_df.drop(columns=[
        f'평당가격_6개월후_minus_{day_offset}',
        f'실거래가_6개월후_minus_{day_offset}',
        f'평당가격_6개월후_gu_minus_{day_offset}',
        f'실거래가_6개월후_gu_minus_{day_offset}',
        '계약일_y'
    ], errors='ignore', inplace=True)
    
    merged_df.rename(columns={'계약일_x': '계약일'}, inplace=True)

    # ✅ NaN 개수 확인 (반복문 실행 후)
    after_nan_price = merged_df['평당가격_6개월후'].isna().sum()
    after_nan_deal = merged_df['실거래가_6개월후'].isna().sum()

    # ✅ 이번 루프에서 채워진 개수 계산
    filled_price = before_nan_price - after_nan_price
    filled_deal = before_nan_deal - after_nan_deal

    if day_offset % 5 == 0 or filled_price > 0 or filled_deal > 0:
        print(f"   ▶ 채우기 후: 평당가격 NaN = {after_nan_price}, 실거래가 NaN = {after_nan_deal}")
        print(f"   ✅ 이번 루프에서 채워진 NaN 개수: 평당가격 {filled_price}개, 실거래가 {filled_deal}개\n")

    # ✅ 변동이 없으면 루프 종료
    if filled_price == 0 and filled_deal == 0:
        print(f"🚨 더 이상 채울 데이터가 없음. 루프 종료 (-{day_offset}일)")
        break

    # ✅ 1일씩 증가
    day_offset += 1

print("✅ NaN 채우기 완료! 🚀")
```

```python
import pandas as pd

day_offset = 1
while merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', ['평당가격_6개월후', '실거래가_6개월후']].isna().any().any():
    # ✅ NaN 개수 확인 (반복문 시작 전)
    before_nan_price = merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', '평당가격_6개월후'].isna().sum()
    before_nan_deal = merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', '실거래가_6개월후'].isna().sum()
    
    if day_offset % 5 == 0 or before_nan_price > 0 or before_nan_deal > 0:
        print(f"🚀 NaN 채우기 진행 중... (-{day_offset}일 데이터 사용)")
        print(f"   ▶ 채우기 전: 평당가격 NaN = {before_nan_price}, 실거래가 NaN = {before_nan_deal}")

    # ✅ 하루 전 날짜 계산
    merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', f'계약일_6개월후_minus_{day_offset}'] = (
        merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', '계약일_6개월후'] - pd.DateOffset(days=day_offset)
    )

    # ✅ 하루 전 날짜를 기준으로 미래 가격 가져오기 (법정동 기준)
    merged_df = pd.merge(merged_df, future_prices, 
                         left_on=['자치구코드', '법정동코드', f'계약일_6개월후_minus_{day_offset}'],
                         right_on=['자치구코드', '법정동코드', '계약일'], how='left', suffixes=('', f'_minus_{day_offset}'))

    # ✅ NaN 값 채우기 (법정동 기준)
    merged_df['평당가격_6개월후'].fillna(merged_df.get(f'평당가격_6개월후_minus_{day_offset}', pd.NA), inplace=True)
    merged_df['실거래가_6개월후'].fillna(merged_df.get(f'실거래가_6개월후_minus_{day_offset}', pd.NA), inplace=True)

    # ✅ 같은 자치구의 하루 전 가격도 가져오기
    merged_df = pd.merge(merged_df, future_prices_gu, 
                         left_on=['자치구코드', f'계약일_6개월후_minus_{day_offset}'],
                         right_on=['자치구코드', '계약일'], how='left', suffixes=('', f'_gu_minus_{day_offset}'))

    # ✅ NaN 값 채우기 (자치구 기준)
    merged_df['평당가격_6개월후'].fillna(merged_df.get(f'평당가격_6개월후_gu_minus_{day_offset}', pd.NA), inplace=True)
    merged_df['실거래가_6개월후'].fillna(merged_df.get(f'실거래가_6개월후_gu_minus_{day_offset}', pd.NA), inplace=True)

    # ✅ 불필요한 컬럼 삭제
    merged_df.drop(columns=[
        f'평당가격_6개월후_minus_{day_offset}',
        f'실거래가_6개월후_minus_{day_offset}',
        f'평당가격_6개월후_gu_minus_{day_offset}',
        f'실거래가_6개월후_gu_minus_{day_offset}',
        '계약일_y'
    ], errors='ignore', inplace=True)
    
    merged_df.rename(columns={'계약일_x': '계약일'}, inplace=True)

    # ✅ NaN 개수 확인 (반복문 실행 후)
    after_nan_price = merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', '평당가격_6개월후'].isna().sum()
    after_nan_deal = merged_df.loc[merged_df['계약일_6개월후'] < '2025-02-13', '실거래가_6개월후'].isna().sum()

    # ✅ 이번 루프에서 채워진 개수 계산
    filled_price = before_nan_price - after_nan_price
    filled_deal = before_nan_deal - after_nan_deal

    if day_offset % 5 == 0 or filled_price > 0 or filled_deal > 0:
        print(f"   ▶ 채우기 후: 평당가격 NaN = {after_nan_price}, 실거래가 NaN = {after_nan_deal}")
        print(f"   ✅ 이번 루프에서 채워진 NaN 개수: 평당가격 {filled_price}개, 실거래가 {filled_deal}개\n")

    # ✅ 변동이 없으면 루프 종료
    if filled_price == 0 and filled_deal == 0:
        print(f"🚨 더 이상 채울 데이터가 없음. 루프 종료 (-{day_offset}일)")
        break

    # ✅ 1일씩 증가
    day_offset += 1

print("✅ NaN 채우기 완료! 🚀")
```
